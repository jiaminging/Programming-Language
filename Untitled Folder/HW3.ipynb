```python
import requests 
import json
import time
from time import sleep
import pandas as pd 
```

```python
case_num=[]
for page in range(1,11):
    urlfirst="https://www.lawplus.com.tw/rest/search/report?querySentence=(%E8%BB%8A%E7%A6%8D%2B%E4%BA%A4%E9%80%9A%E4%BA%8B%E6%95%85)%26(%E6%B0%91%E6%B3%95%E7%AC%AC217%2B%E9%81%8E%E5%A4%B1%E6%AF%94%E4%BE%8B)&keyword=(%E8%BB%8A%E7%A6%8D%2B%E4%BA%A4%E9%80%9A%E4%BA%8B%E6%95%85)%26(%E6%B0%91%E6%B3%95%E7%AC%AC217%2B%E9%81%8E%E5%A4%B1%E6%AF%94%E4%BE%8B)&prevKeyword=(%E8%BB%8A%E7%A6%8D%2B%E4%BA%A4%E9%80%9A%E4%BA%8B%E6%95%85)%26(%E6%B0%91%E6%B3%95%E7%AC%AC217%2B%E9%81%8E%E5%A4%B1%E6%AF%94%E4%BE%8B)&date=2012%2F05%2F16~2022%2F05%2F16&money=&sentence=&caseNum=&caseTypes=&courts=&levels=1&jtypes=&tags=&issue=&main=&judge=&judgeTypes=&lawyer=&litigant=&prosecutor=&clerk=&rows=10&page="
    urlsec="&sortField=&_=1652714083801"
    link=urlfirst+str(page)+urlsec
    res = requests.get(link,verify=False)
    data=json.loads(res.text)
    temp_data=data['rows']

    for i in temp_data:
        case_num.append(i['identifier'])
print(case_num)
```

```python
each_url=[] 
for r in case_num:
    url="https://www.lawplus.com.tw/#reportDetail?id="+r
    each_url.append(url)
for i in each_url:
    print(i)   
```

```python
each_link=[]
for r in case_num:
    each="https://www.lawplus.com.tw/rest/search/report/"+r
    each_link.append(each)
for i in each_link:
    print(i)
```

```python
content=[]

for i in each_link:
    res=requests.get(i,verify=False)
    resquests=json.loads(res.text)
### dictionary存放
    content.append({
    '日期':resquests['response']['reportBase']['identifier'].split(',')[4], 
    '案號':resquests['response']['reportBase']['caseNum'],
    '內文':resquests['response']['reportBase']['content'],
    '法規':resquests['response']['report']['statute'],
    '法官' :resquests['response']['report']['presideJudge'],
    '裁判類型':resquests['response']['report']['type'],
    '案由':resquests['response']['reportBase']['issue']     
})
    sleep(1)
```
